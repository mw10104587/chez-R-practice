<h1>Question D</h1>

```r
# the required function
dcor_all = function(m){
    # the input is a N x P matrix,
    # we want to return a P x P matrix that indicates the cor between each cols
    p = dim(m)[2]
    output = mat.or.vec(p,p)
    
    # loop through dimension p twice so we can find all the combinations
    for(i in 1:p){
        for(j in 1:p){
            output[i, j] = cor(m[,i], m[,j])
        }
    }
    
    return(output)
}
```

```r
# the required test for the function
set.seed(1)
x <- rnorm(500)
df <- sapply(1:4, function(k) x^k + rnorm(500))
dcor_all(df)
```

<h2>E.2 Which Carrier has the highest delay rate</h2>

```r
head(flights)
```

```r
# My assumption here is that, rate of delay = (delay count)/(total departure count)

# a function that takes a splitted matrix as input
# and calculates the delay rate of this carrier
delay_ratio <- function(x){
    
    # take only the dep_delay column
    x = x$dep_delay
    
    # count the total length of column, which is the amoun of departure for this carrier
    total_dep_count = length(x)
    
    # count the length of this column after filtering out the ones that are smaller or equal to 0,
    # because when the delay time is smaller or equal to 0, it means there wasn't a delay
    delay_count = length(x[x>0])
    
    # return the ratio
    return(delay_count/total_dep_count)
}

# split the flights data according to carrier
split_carrier <- split(flights, flights$carrier)

# apply the function to each split
delay_rate <- sapply(split_carrier, delay_ratio)

# sort the answer and look at the output and learned that
delay_rate[order(delay_rate)]

# WN has the highest delay rate: 0.5498
```

```r
# by using the unique function, we'll only have one instance for each airplane
# by counting the length of the result, we'll have the number of unique airplanes.
length(unique(flights$tailnum))
```

```r
# by using the same split and apply function here,
# we split according to the tailnum
split_tailnum <- split(flights, flights$tailnum)

avg_speed <- function(airplane){
    # input is a table for a unique airplane
    
    # remove data with any part of calculation that has a NA in it
    airplane = airplane[!is.na(airplane$hour),]
    airplane = airplane[!is.na(airplane$minute),]
    airplane = airplane[!is.na(airplane$distance),]
    
    trip <- length(airplane)
    hours <- sum(airplane$hour)
    minutes <- sum(airplane$minute)
    
    total_distance <- sum(airplane$distance)
    
    # the average speed in the unit of (km/hr)
    return(total_distance/(hours + (minutes/60) ))
}

average_speed = sapply(split_tailnum, avg_speed)
average_speed[order(average_speed, decreasing=TRUE)]

# by sorting the result, we can see that N395HA has the largest average speed.
```
